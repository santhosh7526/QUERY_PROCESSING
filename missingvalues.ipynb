```python
import numpy as np
import pandas as pd

df = pd.read_csv('/content/laptopData-selected-columns.csv')

if 'Unnamed: 0' in df.columns:
    df.drop('Unnamed: 0', axis=1, inplace=True)

if df.duplicated().sum() > 0:
    df.drop_duplicates(inplace=True)

df['Inches'] = pd.to_numeric(df['Inches'], errors='coerce')

if df['Inches'].isnull().any():
    df['Inches'].fillna(df['Inches'].mode()[0], inplace=True)

df['Ram'] = df['Ram'].str.replace('GB', '', regex=False)
df['Ram'] = pd.to_numeric(df['Ram'], errors='coerce')

if df['Ram'].isnull().any():
    df['Ram'].fillna(df['Ram'].mode()[0], inplace=True)

df['Ram'] = df['Ram'].astype(int)

def parse_memory(memory_str):
    if pd.isna(memory_str):
        return 0, "Unknown"

    memory_str = str(memory_str)

    memory_gb = 0

    if '+' in memory_str:
        parts = memory_str.split('+')
    else:
        parts = [memory_str]

    storage_types = []

    for part in parts:
        part = part.strip()

        if 'TB' in part:
            size = float(part.split('TB')[0].strip())
            memory_gb += size * 1024
        elif 'GB' in part:
            size = float(part.split('GB')[0].strip())
            memory_gb += size

        if 'SSD' in part:
            storage_types.append('SSD')
        elif 'HDD' in part:
            storage_types.append('HDD')
        elif 'Flash Storage' in part:
            storage_types.append('Flash Storage')
        elif 'Hybrid' in part:
            storage_types.append('Hybrid')

    storage_type = '+'.join(storage_types) if storage_types else "Unknown"

    return memory_gb, storage_type
display(df.head())
```
